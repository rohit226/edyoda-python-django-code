```python
class Student:
    def __init__(self,roll_no,name):
        self.__roll_no = roll_no
        self.__name = name
        
    def get_roll_no(self):
        return self.__roll_no
        
    def set_roll_no(self,new_roll_no):
        self.__roll_no = new_roll_no
        
    def get_name(self):
        return self.__name
        
    def set_name(self,new_name):
        self.__name = new_name
        
    def add_marks(self,marks):
        self.__marks = marks
        
    def get_marks(self):
        return self.__marks
        
    def set_marks(self,marks):
        self.__marks = marks
        
        
    def cal_result(self):
        self.per = sum(self.__marks.values())/500 * 100
        if self.per >= 40:
            self.result = "Pass"
        else:
            self.result = "Fail"
            
    def get_details(self):
        return "{} {} {} {}".format(str(self.__roll_no),self.__name,round(self.per,2),self.result)
    
    
stud1 = Student(1,"ABC")
stud2 = Student(2,"PQR")
stud3 = Student(3,"RST")
stud4 = Student(4,"XYZ")
stud5 = Student(5,"PPP")
stud6 = Student(6,"TTT")

stud1.add_marks({"eng":78,"maths":88,"his":88,"sci":88,"geo":99})
stud2.add_marks({"eng":80,"maths":77,"his":89,"sci":70,"geo":40})
stud3.add_marks({"eng":50,"maths":33,"his":30,"sci":30,"geo":30})
stud4.add_marks({"eng":30,"maths":44,"his":35,"sci":56,"geo":30})
stud5.add_marks({"eng":20,"maths":20,"his":20,"sci":44,"geo":20})
stud6.add_marks({"eng":35,"maths":66,"his":89,"sci":76,"geo":99})

students = [stud1,stud2,stud3,stud4,stud5,stud6]
for student in students:
    student.cal_result()
    

#Failed Students
for student in students:
    if student.result == "Fail":
        print(student.get_details())
        
student_per = [(student,student.per) for student in students if student.result == "Pass"]
# print(student_per)
topper = max(student_per,key = lambda t : t[1])
# print(topper)
# print("Topper of the class is",topper[0].get_detail
print(stud1.set_name("xzc"))
print(stud1.get_name())
print(stud2.set_marks("52"))
print(stud2.get_marks())
print(stud1.get_marks())
```
